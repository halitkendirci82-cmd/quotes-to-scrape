```
import requests
from bs4 import BeautifulSoup

quotes = []

for page in range(1, 11):
    url = f"http://quotes.toscrape.com/page/{page}/"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    quote_tags = soup.find_all("span", class_="text")

    for q in quote_tags:
        quotes.append(q.text)

print(len(quotes))
```

```
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(quotes)

words = vectorizer.get_feature_names_out()
```

```
import numpy as np

word_counts = X.sum(axis=0)
word_counts = np.array(word_counts).flatten()

freq = dict(zip(words, word_counts))

top5 = sorted(freq.items(), key=lambda x: x[1], reverse=True)[:5]

print(top5)
```

```
import requests
from bs4 import BeautifulSoup
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

quotes = []

for page in range(1, 11):
    url = f"http://quotes.toscrape.com/page/{page}/"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    quote_tags = soup.find_all("span", class_="text")

    for q in quote_tags:
        quotes.append(q.text)

vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(quotes)

words = vectorizer.get_feature_names_out()

df = pd.DataFrame(X.toarray(), columns=words)
df[df > 0] = 1

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

display(df.head())
```

무엇인가?

----

1. 단어의 의미(semantic)를 이해하지 못한다.
2. 단어의 순서를 고려하지 않는다.
3. 차원이 매우 커질 수 있다 (컬럼 수가 많아짐).
